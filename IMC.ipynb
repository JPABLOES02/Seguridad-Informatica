```python
#Funcion que determine IMC y niveles de peso 
datos = {
    "nombre": "Pablo Espinosa",
    "peso": 75,
    "altura": 1.75  
}

def imc(peso, altura, nombre):
    imc_p = peso / (altura**2)  
    
    if imc_p < 18.5:
        print(f"{nombre}, est치s bajo de peso:Tu IMC es de: IMC {round(imc_p, 1)}")
    elif imc_p >= 18.5 and imc_p <= 24.9:
        print(f"{nombre}, est치s en un rango de peso normal:Tu IMC es de: {round(imc_p, 1)}")
    elif imc_p >= 25.0 and imc_p <= 29.9:
        print(f"{nombre}, est치s en sobrepeso:Tu IMC es de: IMC {round(imc_p, 1)}")
    else:
        print(f"{nombre}, tienes obesidad:Tu IMC es de: IMC {round(imc_p, 1)}")

peso = datos["peso"]
altura = datos["altura"]
nombre = datos["nombre"]

imc(peso, altura, nombre)
```
